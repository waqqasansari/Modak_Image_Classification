```python
from google_images_download import google_images_download
```

```python
response = google_images_download.googleimagesdownload()

search_queries = [ 
  'barfi',
    'kaju katli',
    'gajar ka halwa',
    'indian sweets',
    'laddu',
    'kheer',
    'jalebi',
    'gulab jamun',
    'rossogulla',
    'aflatoon dessert'
] 

```

```python
def downloadimages(query):
    arguments = {"keywords": query, 
                 "format": "jpg", 
                 "limit":50, 
                 "print_urls":True, 
                 "size": "medium", 
                 "aspect_ratio": "square"}
    try:
        response.download(arguments)
    except FileNotFoundError:
        arguments = {"keywords": query, 
                     "format": "jpg", 
                     "limit":4, 
                     "print_urls":True,  
                     "size": "medium"}
        try:
            response.download(arguments)
        except:
            pass
```

```python
for query in search_queries: 
    downloadimages(query)  
    print()  
```
